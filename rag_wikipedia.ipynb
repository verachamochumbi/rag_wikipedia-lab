```python
import pandas as pd
import textwrap

df = pd.read_csv("data/wiki_corpus.csv")

text = df.loc[0, "text"]

# Dividir en chunks de 300 palabras
words = text.split()
chunk_size = 300
chunks = [" ".join(words[i:i+chunk_size]) for i in range(0, len(words), chunk_size)]

print(f"Total chunks creados: {len(chunks)}")
```

```python
from sentence_transformers import SentenceTransformer
import chromadb

# Crear cliente ChromaDB
client = chromadb.Client()
collection = client.create_collection("wiki_ai")

# Embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Subir chunks a ChromaDB
for i, chunk in enumerate(chunks):
    embedding = model.encode(chunk).tolist()

    collection.add(
        ids=[str(i)],
        embeddings=[embedding],
        metadatas=[{"source": "Federated_Learning"}],
        documents=[chunk]
    )

print("Embeddings guardados en ChromaDB ✔️")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

model_name = "facebook/bart-large-cnn"

tokenizer = AutoTokenizer.from_pretrained(model_name)
hf_model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

summarizer = pipeline(
    "summarization",
    model=hf_model,
    tokenizer=tokenizer,
    max_length=260,
    min_length=80,
    do_sample=False,
)
```

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import SentenceTransformerEmbeddings

embedding = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

vectordb = Chroma(
    collection_name="wiki_ai",
    embedding_function=embedding,
)

retriever = vectordb.as_retriever(search_kwargs={"k": 4})
```

```python
def answer_query(query: str, k: int = 4) -> str:
    # 1. Recuperar docs relevantes
    docs = retriever.invoke(query)  # invoke en vez de get_relevant_documents
    if not docs:
        return "No se encontraron documentos relevantes."

    # 2. Combinar el contenido en un solo contexto
    context = "\n\n".join([d.page_content for d in docs])

    # 3. TRUNCAR EL CONTEXTO A 2000 CARACTERES (BART no acepta más)
    context = context[:2000]

    # 4. Construir prompt
    prompt = (
        "Context:\n"
        + context
        + "\n\nQuestion: "
        + query
        + "\n\nAnswer in a clear, concise paragraph:"
    )

    # 5. Generar respuesta usando el summarizer
    out = summarizer(
        prompt,
        max_length=250,
        min_length=80,
        do_sample=False,
    )[0]["summary_text"]

    return out
```

```python
# 1. Recuperar documentos relevantes con invoke
docs = retriever.invoke(
    "Challenges of applying federated learning in healthcare."
)
print(f"Se recuperaron {len(docs)} documentos.")

# 2. Combinar el contenido de esos documentos
combined_text = "\n\n".join([d.page_content for d in docs])

# 3. TRUNCAR (fundamental para evitar errores de BART)
combined_text = combined_text[:2000]

# 4. Resumir con el summarizer
rag_summary = summarizer(
    combined_text,
    max_length=260,
    min_length=120,
    do_sample=False,
)[0]["summary_text"]

print(rag_summary[:500])
```

```python
import os

os.makedirs("outputs", exist_ok=True)

md_path = "outputs/rag_summary.md"
with open(md_path, "w", encoding="utf-8") as f:
    f.write("# RAG Summary: Federated Learning in Healthcare\n\n")
    f.write(rag_summary)

md_path
```

```python
import json

queries = [
    "Explain federated learning in simple terms.",
    "What are the key privacy benefits of federated learning?",
    "What are the main challenges of federated learning in healthcare?",
]

results = []
for q in queries:
    try:
        ans = answer_query(q)
        results.append({"query": q, "answer": ans})
    except Exception as e:
        results.append({"query": q, "error": str(e)})

json_path = "outputs/retrieval_examples.json"
with open(json_path, "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=2)

json_path
```
